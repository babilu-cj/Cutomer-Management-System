###### Q. Print all shipments which are delivered within 7 days of courier Start date

```python
from datetime import datetime as dt
```

```python
for k,v in shipment.items():
    if v[3]=="Null":
        continue
    else:
        ship_date=dt.strptime(v[2],"%d-%m-%Y")
        deliv_date=dt.strptime(v[3],"%d-%m-%Y")
        if (deliv_date-ship_date).days<=7:
            print("Shipment ID:",k,"\b, Details: ",v)
```

###### Q. Print all shipments which are delivered after 15 days of courier start date or not yet been delivered.

```python
for k,v in shipment.items():
    if v[3]=="Null":
        print("Shipment ID:",k,"\b, Remarks: Not Yet Delivered")
        print("Details: ",v,"\n")
    else:
        ship_date=dt.strptime(v[2],"%d-%m-%Y")
        deliv_date=dt.strptime(v[3],"%d-%m-%Y")
        if (deliv_date-ship_date).days>15:
            print("Shipment ID:",k,"\b, Remarks: Delivered After 15 Days")
            print("Details: ",v,"\n")
        else:
            continue
```
